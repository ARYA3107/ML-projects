## Selecting the Dependent(target) and Independent Variables:

# Spliting the data into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

# Models:

## Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn import metrics
from sklearn.model_selection import cross_val_score
```

```python
param_grid={'max_depth': [0.5, 1, 5, 10],
                         'n_estimators': [16, 32, 50, 100], 'n_jobs': [1, 2],
                         'random_state': [1, 10, 20, 42]}
```

```python
gsv=GridSearchCV(RandomForestClassifier(),param_grid,cv=3,scoring='accuracy')
```

```python
gsv.fit(X_train,y_train)
```

```python
RandomForestClassifier()
```

```python
print('Best estimator is:',gsv.best_estimator_)
print('Best parameters are:',gsv.best_params_)
print('Best score is:',gsv.best_score_)
```

```python
model=RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=10)
```

```python
model.fit(X_train,y_train)
```

### Checking  accuracy score, Confusion metrics, Classification report and Cross val score

```python
model.score(X_train,y_train)
```

```python
y_pred=model.predict(X_test)
```

```python
print('Accuracy score is',metrics.accuracy_score(y_test,y_pred))
print('Confusion metrics:')
print(metrics.confusion_matrix(y_test,y_pred))
print('Classification report:')
print(metrics.classification_report(y_test,y_pred))
print('Cross val score:',cross_val_score(RandomForestClassifier(),X,y,cv=3).mean())
```

```python
def getprediction():
    sex1=sex.get()
    age1=age.get()
    job1=job.get()
    housing1=housing.get()
    Savingaccounts1=saving.get()
    Checkingaccount1=checking.get()
    Creditamount1=credit.get()
    duration1=d.get()
    purpose1=p.get()
    
    
    if sex1=='Male':
            sex1= 1
    else:
        sex1= 0
        
    if housing1=='own':
        housing1=1
    elif housing1=='free':
        housing1=0
    else:
        housing1=2
        
    if Savingaccounts1=='unknown':
        Savingaccounts1=4
    elif Savingaccounts1=='little':
        Savingaccounts1=0
    elif Savingaccounts1=='quite rich':
        Savingaccounts1=2
    elif Savingaccounts1=='rich':
        Savingaccounts1=3
    else:
        Savingaccounts1=1
        
    if Checkingaccount1=='little':
        Checkingaccount1=0
    elif Checkingaccount1=='moderte':
        Checkingaccount1=1
    elif Checkingaccount1=='rich':
        Checkingaccount1=3
    else:
        Checkingaccount1=2 
            
    
    if purpose1=='radio/TV':
        radio=1
        education=0
        furniture=0
        car=0
        business=0
        domestic=0
        repairs=0
        vacation=0
    elif purpose1=='education':
        radio=0
        education=1
        furniture=0
        car=0
        business=0
        domestic=0
        repairs=0
        vacation=0
    elif purpose1==' furniture/equipment':
        radio=0
        education=0
        furniture=1
        car=0
        business=0
        domestic=0
        repairs=0
        vacation=0
    elif purpose1=='car':
        radio=0
        education=0
        furniture=0
        car=1
        business=0
        domestic=0
        repairs=0
        vacation=0
    elif purpose1=='business':
        radio=0
        education=0
        furniture=0
        car=0
        business=1
        domestic=0
        repairs=0
        vacation=0 
    elif purpose1=='domestic appliances':
        radio=0
        education=0
        furniture=0
        car=0
        business=0
        domestic=1
        repairs=0
        vacation=0
    elif purpose1=='repairs':
        radio=0
        education=0
        furniture=0
        car=0
        business=0
        domestic=0
        repairs=1
        vacation=0
    else:
        radio=0
        education=0
        furniture=0
        car=0
        business=0
        domestic=0
        repairs=0
        vacation=1
        
    x=scaler.transform([[age1,sex1,job1,housing1,Savingaccounts1,Checkingaccount1,Creditamount1,duration1,radio,education,furniture,car,business,domestic,repairs,vacation]])
    prediction=model2.predict(x)
    
    result.delete(0,END)
    
    if prediction[0]==0:
        result.insert(0,'Good') 
    else:
        result.insert(0,'Bad')
        
```

```python
window=Tk()
window.title('German Credit Risk Analysis')
window.geometry('600x800')
titlelabel=Label(window,text='German Credit Risk Analysis',bg='pink')
titlelabel.pack(fill='both')

Label(window).pack()
agelabel=Label(window,text='Age').pack()
age=Entry(window)
age.pack()

Label(window).pack()
sexlabel=Label(window,text='Sex').pack()
sex=Combobox(window)
sex['values']=('Male','Female')
sex.current(0)
sex.pack()

Label(window).pack()
joblabel=Label(window,text='job').pack()
job=Combobox(window)
job['values']=('2','1','3','0')
job.current(0)
job.pack()

Label(window).pack()
housinglabel=Label(window,text='Housing').pack()
housing=Combobox(window)
housing['values']=('own','free','rent')
housing.current(0)
housing.pack()

Label(window).pack()
Savingaccountslabel=Label(window,text='Saving accounts').pack()
saving=Combobox(window)
saving['values']=('unknown','little', 'quite rich', 'rich', 'moderate')
saving.current(0)
saving.pack()

Label(window).pack()
Checkingaccountlabel=Label(window,text='Checking account').pack()
checking=Combobox(window)
checking['values']=('little', 'moderate','unknown', 'rich')
checking.current(0)
checking.pack()

Label(window).pack()
Creditamountlabel=Label(window,text='Credit amount').pack()
credit=Entry(window)
credit.pack()

Label(window).pack()
durationlabel=Label(window,text='Duration').pack()
d=Entry(window)
d.pack()

Label(window).pack()
Purposelabel=Label(window,text='Purpose').pack()
p=Combobox(window)
p['values']=('radio/TV','education','furniture/equipment','car','business','domestic appliances','repairs','vacation/others')
p.current(0)            
p.pack()

Label(window).pack()
button=Button(window,text='Get Prediction',command=getprediction)
button.pack()
result=Entry(window)
result.pack()

window.mainloop()
```
